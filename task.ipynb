````python
# read the trainings data
dtrain = pd.read_csv('./data/flutype_train.csv', sep="\t")
dtrain
````

````python
# build classifier with trainings dataset
from sklearn.linear_model import LogisticRegression
X_train = dtrain.ix[:, dtrain.columns != 'virus']
Y_train = dtrain['virus']

clf = LogisticRegression()
clf = clf.fit(X_train, Y_train)

# save classifier to disk
pd.to_pickle(clf, "results/logistic_regressor.dat")
````

````python
# run classifier on test dataset
X_test = dtest.ix[:, dtest.columns != 'virus']
Y_test = dtest['virus'] 
predicted_test = clf.predict(X_test) 

# run classifier on training dataset
predicted_train = clf.predict(X_train) 
````

# Evaluation

````python
from  sklearn.metrics import *
# convert predicted and true classes to list
predicted_train = list(predicted_train)
predicted_test = list(predicted_test)
Y_train = list(Y_train)
Y_test = list(Y_test)

# look at classification report showing performance 
# for each class separatedly
report = classification_report(Y_test, predicted_test)
print report
````

````python
# generate summary table containing some evaluation metrices for the 
# performance of the classifier on traing set and on test set
test_set_eval = [accuracy_score(predicted_test, Y_test),
                 recall_score(predicted_test, Y_test, average='micro'),
                 precision_score(predicted_test, Y_test, average='micro'),
                 f1_score(predicted_test, Y_test, average='micro')]

training_set_eval = [accuracy_score(predicted_train, Y_train),
                     recall_score(predicted_train, Y_train, average='micro'),
                     precision_score(predicted_train, Y_train, average='micro'),
                     f1_score(predicted_train, Y_train, average='micro')]

result_table_columns = ["Accuracy", "Recall", "Precission", "F-Score"]
result_table = pd.DataFrame({ 'Meassure' : result_table_columns,
                              'Training Set' : training_set_eval,
                              'Test Set' : test_set_eval})

# save table to disk
result_table.to_csv("results/result_table.csv",  sep="\t")
result_table
````

````python
%matplotlib inline
# define function to plot confusion matrix 
import itertools
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, plt_name, classes, title="Confusion Matrix"):
    tick_marks = range(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(title)
    plt.colorbar()
    plt.ylabel('True Class')
    plt.xlabel('Predicted Class')
    plt.savefig(plt_name)
    plt.show()
    plt.close()

# plot confusion matrix for predicitions on training set
class_names = dtest.virus.unique().tolist()
cnf_matrix_train = confusion_matrix(Y_train, predicted_train, labels=class_names)
plot_confusion_matrix(cnf_matrix_train,
                      plt_name = "results/confusion_matrix_training_set.png",
                      classes=class_names,
                      title = "Confusion Matrix (training set)")

# plot confusion matrix for predicitions on test set
cnf_matrix_test = confusion_matrix(Y_test, predicted_test, labels=class_names)
plot_confusion_matrix(cnf_matrix_test,
                      plt_name = "results/confusion_matrix_test_set.png",
                      classes=class_names,
                      title = "Confusion Matrix (test set)")
````

````python
%matplotlib inline
# alternatively train a SVM and evaluat it via the ROC-AUC
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from sklearn import svm

# binarize the output
Y_test_bin = label_binarize(list(Y_test), classes=class_names)
Y_train_bin = label_binarize(list(Y_train), classes=class_names)
n_classes = len(class_names)

# learn to predict each class against the other
classifier = OneVsRestClassifier(svm.SVC(kernel='linear', probability=True))
y_score = classifier.fit(X_train, Y_train_bin).decision_function(X_test)

# compute micro-average ROC curve and ROC-AUC
fpr, tpr, _ = roc_curve(Y_test_bin.ravel(), y_score.ravel())
roc_auc = auc(fpr, tpr)

# plot ROC curve
plt.figure()
plt.plot(fpr, tpr, label='micro-avg. ROC curve, AUC = %s' %roc_auc, linewidth=8)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.savefig('results/multi-class_roc.png')
````
